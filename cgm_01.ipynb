```python
def array(xs):
  return t.VentureArrayUnboxed(np.array(xs),  t.NumberType())

def makeObservations(x,y,gp_str='(gp '):
    xString = genSamples(x,gp_str)
    ripl.observe(xString, array(y))

def genSamples(x,gp_str='(gp '):
    sampleString=gp_str+' (array '
    for i in range(len(x)):
        sampleString+= str(x[i]) + ' '
    sampleString+='))'
    #print(sampleString)
    return sampleString
```

```python
ripl = shortcuts.make_lite_church_prime_ripl()
ripl.bind_foreign_sp("make_gp_part_der",gp_der.makeGPSP)
ripl.assume('make_const_func', VentureFunction(makeConstFunc, [t.NumberType()], constantType))
ripl.assume('zero', "(apply_function make_const_func 0)")
```

```python
ripl.assume("func_times", makeLiftedMult(lambda x1, x2: np.multiply(x1,x2)))
ripl.assume("func_plus", makeLiftedAdd(lambda x1, x2: x1 + x2))
```

```python
ripl.assume('make_periodic', VentureFunction(makePeriodic, [t.NumberType(), t.NumberType(), t.NumberType()], t.AnyType("VentureFunction")))
ripl.assume('make_se',VentureFunction(makeSquaredExponential,[t.NumberType(), t.NumberType()], t.AnyType("VentureFunction")))
ripl.assume('make_noise', VentureFunction(makeNoise, [t.NumberType()], t.AnyType("VentureFunction")))
ripl.assume('make_rq',VentureFunction(makeRQ, [t.NumberType(), t.NumberType(), t.NumberType()], t.AnyType("VentureFunction")))
ripl.assume('make_linear', VentureFunction(makeLinear, [t.NumberType()], t.AnyType("VentureFunction")))


'''
ripl.assume('sf1','(tag (quote hyper) 0 (log (gamma 3 2)))')
ripl.assume('l1','(tag (quote hyper) 1 (log (gamma 3 2)))')



ripl.assume('ell','(tag (quote hyper) 2 (log (gamma 3 2)))')
ripl.assume('p','(tag (quote hyper) 3 (log (gamma 3 2)))')
ripl.assume('s','(tag (quote hyper) 4 (log (gamma 3 2)))')
'''

ripl.assume('sf1','(tag (quote hyper) 0 (log (uniform_continuous 0 5)))')
ripl.assume('l1','(tag (quote hyper) 1 (log (uniform_continuous 0 5)))')

ripl.assume('sf2','(tag (quote hyper) 9 (log (uniform_continuous 0 5)))')
ripl.assume('l2','(tag (quote hyper) 10 (log (uniform_continuous 0 5)))')





ripl.assume('ell','(tag (quote hyper) 2 (log (uniform_continuous 0 5)))')
ripl.assume('p','(tag (quote hyper) 3 (log (uniform_continuous 0 5)))')
ripl.assume('s','(tag (quote hyper) 4 (log (uniform_continuous 0 5)))')

ripl.assume('a1','(tag (quote hyper) 5 (log (uniform_continuous 0 5)))')
#ripl.assume('a2','(tag (quote hyper) 6 (log (uniform_continuous 0 10)))')

ripl.assume('a','(tag (quote hyper) 6 (log (uniform_continuous 0 5)))')
ripl.assume('b','(tag (quote hyper) 7 (log (uniform_continuous 0 5)))')
ripl.assume('c','(tag (quote hyper) 8 (log (uniform_continuous 0 5)))')
```

```python
ripl.assume('se1', "(apply_function make_se sf1 l1 )")
ripl.assume('se2', "(apply_function make_se sf2 l2 )")
ripl.assume('per', "(apply_function make_periodic ell p s)")

ripl.assume('lin1', "(apply_function make_linear a1 )")
#ripl.assume('lin2', "(apply_function make_linear a2 )")

ripl.assume('rq', "(apply_function make_rq a b c)")
```

```python
ripl.assume('gp',"""(tag (quote model) 0
                        (make_gp_part_der zero
                            (apply_function func_plus
                                (apply_function func_times se1 
                                    (apply_function func_plus lin1 per)
                                )
                                (apply_function func_times se2 rq)
                                
                            )
                         )
                             )""")
```
